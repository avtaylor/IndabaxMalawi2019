```python
#import the data which in this case is a csv file containing 100 
# common words in Chichewa in Chichewa and with their English translation
#Source: https://1000mostcommonwords.com/1000-most-common-chichewa-words/
# You can read more about Python lists for example here:
# https://www.programiz.com/python-programming/list
#For this we need to import the Python library for reading text files
import csv

with open('100WordsChichewa.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            print(f'\t{row[0]} Chichewa: {row[1]} English: {row[2]}.')
            line_count += 1
    print(f'Processed {line_count} lines.')
```

```python
# The code above simply prints the contents of the csv file
# in some sort of template. 
# What we want is to hold this data in a variable and use it
# We can use lists
# So most of teh code above stays as it is except that instead of
# printing out the words we store them into a list
# The list is called  wordsList
# We use wordsList.append(row) to append each row in the file to the list
with open('100WordsChichewa.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    wordsList=[]
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            wordsList.append(row)
            line_count += 1
    print(f'Processed {line_count} lines.')
```

```python
# Operations such as adding, deleting, replacing elements is possible
# Lets say we want to add the noun ndiwo
# First we need to check if it is in the list
w="ndiwo"
found=False
for row in wordsList:
    if w == row[1]:
        found=True
        break
if found:
    print("%s is already in the list" %w)
else:
    print("%s is not in the list" %w)
    # so we can add it
    wordsList.append(w) 
```

```python
# Suppose I want to re-use this code several times to check for
# words if they are or not in our list
# We can define a function 
def inList(w,wordsList,EngTranslation):
    found=False
    for row in wordsList:
        if w == row[2]:
            found=True
            break
    if found:
        print("%s is already in the list" %w)
        return True
    else:
        print("%s is not in the list so will be added" %w )
        # so we can add it in the same format in which the others 
        # are in teh list so we need the counter and teh English translation
        wordsList.append([len(wordsList)+1,w,EngTranslation])
        return False
```

```python
# Let us create a list that contains all the letters of the English alphabet.
# We will do this the beautiful python was and use the map funtion
# Map takes a function and an iterable as arguments. 
# map applies this function to every item of the iterable and 
# returns a list.

alphabet = map(chr, range(97, 123))
for x in alphabet:
    print(x, end=' ')
#dir(alphabet)
```

```python
# LAMBDAS are frequently use in Python 
# these are like definitions of functions which 
# can be used as delayed executions
# For example i defined a lambda function with two parameters x, y
# which is supposed to check if y appears in x on the first position.
starts_with= lambda x,y: y in x[0]
print(starts_with('ntchito','n'))
```
