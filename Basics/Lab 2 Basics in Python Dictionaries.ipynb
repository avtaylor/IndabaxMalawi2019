```python
# We are going to put the csv file from the List example
# in a data structure called dictionary
# See more on dict here: 
#https://www.programiz.com/python-programming/dictionary
import csv
with open('100WordsChichewa.csv') as csv_file:
    reader=csv.reader(csv_file)
    wordsList = {rows[1]:rows[2] for rows in reader}
wordsList
```

```python
#How to access elements in a dict
wordsList['Chichewa']
```

```python
wordsList.pop("Chichewa")#this corresponds to the first row in the csv file
# so we can remove it
```
